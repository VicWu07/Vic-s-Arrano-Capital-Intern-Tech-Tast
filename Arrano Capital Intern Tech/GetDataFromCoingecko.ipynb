# Below is the Process of finding top 80 ERC-20

This File is used to Fetch Needed data using Coingecko API to store into a .db file

```python
import sqlite3 as sql
import pandas as pd
import matplotlib.pyplot as plt
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
coins=cg.get_coins_list(include_platform=True)
df_coins=pd.DataFrame(coins, columns=['name','id','symbol','platforms'])
df_coins #Table Below get a list of coins from Coingecko API
```

```python
markets=cg.get_coins_markets(vs_currency="usd",order="market_cap_desc",per_page=250)
markets #the TOP 250 coins by market cap in desceding order
```

```python
df_markets=pd.DataFrame(markets,columns=['id','symbol','market_cap_rank','name'])
df_top80 = pd.read_csv("GetDataFromCoingecko.csv",usecols=[2],nrows=80)
ls_1 = df_top80['Column 3'].tolist()
ls_2 = []
for strtemp in ls_1:
    ls_2.append(strtemp.split('\n')[0])
df_top80_markets=df_markets[df_markets['name'].isin(ls_2)]
ls_id = df_top80_markets['id'].tolist() #get the id from scaped csv from coingecko
token_add = df_coins[df_coins['id'].isin(ls_id)]
token_add #Compare the list of coins to id of ERC tokens from csv doc, in order to find address and properties of ERC20
```

```python
ls_temp1=token_add['platforms'].tolist()
ls_add= [d.get('ethereum')for d in ls_temp1] #get address from dictionary
```

```python
token_add1 = token_add.drop(columns=['platforms'])
token_add1['token_address'] = ls_add
token_add1 #the list of ERC20 ranked top80 with token address
```

```python
df_top80_markets
```

```python
df_top80_markets=df_top80_markets.set_index('id')
token_add1=token_add1.set_index('id')
#above action set index of both dataframe as 'id' for further merge
```

```python
token_add1
```

```python
df_top80_markets['id']=df_top80_markets.index
df_preSQL=df_top80_markets.merge(token_add1,how='left') #left merge the table in order to save into SQL basicinfo table
```

```python
df_preSQL
```

# Below is the Process of Saving DataFrame to SQLite

```python
con = sql.connect("Top80ERC20.db")

cur = con.cursor()
df_preSQL.to_sql("ERC20Basic",con,if_exists='replace')
con.commit()
con.close()
```
